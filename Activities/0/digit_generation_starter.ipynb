```python
import numpy as np
import matplotlib.pyplot as plt

# create a matrix to represent a drawing of '7'
x = np.zeros((8,8))
x[2,3:6] = 1
x[3,5] = 1
x[4:6,4] = 1

# display image
plt.imshow(np.round(x,0), cmap='gray_r'), plt.axis('off')
plt.show()
```
