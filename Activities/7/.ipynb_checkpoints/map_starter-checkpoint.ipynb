```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5,10,100)

mu1 = 0
var1 = 1
pdf1 = norm.pdf(x, mu1, np.sqrt(var1))

mu2 = 3
var2 = 1
pdf2 = norm.pdf(x, mu2, np.sqrt(var2))

plt.plot(x,pdf1)
plt.plot(x,pdf2)

plt.xlabel('x')
plt.show()
```
