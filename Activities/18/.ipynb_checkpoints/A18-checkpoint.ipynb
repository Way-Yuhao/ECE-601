```python
import numpy as np
from sklearn import mixture
import math
```

```python
def EM(xs, K, n):
    # initalization
    l = len(xs)
    mus = np.array([-2, 2])
    pis = np.array([.5, .5])
    sigmas = np.array([[1], [1]])
    r = np.zeros(l)
    for _ in range(n):
        # E-Step
        for i in range(l):
            if abs(xs[i] - mus[0]) >= abs(xs[i] - mus[1]):
                r[i] = 0
            else:
                r[i] = 1
        # M-Step
        mus[0] = np.mean(xs[r == 0])
        mus[1] = np.mean(xs[r == 1])
        sigmas[0] = 1/np.count_nonzero(r) * np.add(xs[r == 0], -mus[0]) @ np.add(xs[r == 0], - mus[0]).T
        sigmas[1] = 1/(l-np.count_nonzero(r)) * np.add(xs[r == 1], - mus[1]) @ np.add(xs[r == 1], - mus[1]).T
        pis = np.array([np.count_nonzero(r)/l, (l-np.count_nonzero(r))/l])
        print("=========")
        print("mu = ", mus)
        print("sigma = ", sigmas)
        print("pi = ", pis)
        print("=========")
```

```python

D = np.array([-9, -8, -7, 5, 7, 9])
EM(D, 2, 10)
```
